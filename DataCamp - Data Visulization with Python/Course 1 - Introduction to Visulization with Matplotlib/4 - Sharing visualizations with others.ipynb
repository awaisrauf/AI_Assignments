## Preparing Figures to Share with Other

````python
import pandas as pd
import matplotlib.pyplot as plt

# using dates as index, two ways
climate_data = pd.read_csv("data/climate_change.csv", index_col='date', parse_dates=True)
climate_data.index = pd.to_datetime(climate_data.index)
````

````python
fig, ax = plt.subplots()
ax.plot(climate_data.index, climate_data['co2'])
ax.plot(climate_data.index, climate_data['co2']+10)
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
````

### Choosing a Style

We can change sytle of the plot figures by using `plt.style.use('name-of-style)`
1. ggplot sytle
2. bmh 
3. seaborn-colorblind
4. gray scale
5. Solarize_Light2

````python
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.plot(climate_data.index, climate_data['co2'])
ax.plot(climate_data.index, climate_data['co2']+10)
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
````

````python
plt.style.use('bmh')
fig, ax = plt.subplots()
ax.plot(climate_data.index, climate_data['co2'])
ax.plot(climate_data.index, climate_data['co2']+10)
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
````

````python
plt.style.use('seaborn-colorblind')
fig, ax = plt.subplots()
ax.plot(climate_data.index, climate_data['co2'])
ax.plot(climate_data.index, climate_data['co2']+10)
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
````

````python
plt.style.use('grayscale')
fig, ax = plt.subplots()
ax.plot(climate_data.index, climate_data['co2'])
ax.plot(climate_data.index, climate_data['co2']+10)
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
````

````python
plt.style.use('Solarize_Light2')
fig, ax = plt.subplots()
ax.plot(climate_data.index, climate_data['co2'])
ax.plot(climate_data.index, climate_data['co2']+10)
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
````

### Saving Visulizations

to save a figure:
```python
plt.savefig('name.png', kwargs)
```
1. `.png` is lossless compression but takes more disk space
2. `jpeg` is lossy but takess less space
3. `.svg` if you want the figure to edit in adobe or similar software
3. `quality` how small the file and degree of loss of quality; >95 means compression does not work
4. `dpi`: dots per inch, higher the number higher resolution and larger file size
5. `fig.set_size_inches([5, 3])`: to change the size i.e. aspect ratio `[width, height]`


````python
# loading data
import pandas as pd
medals = pd.read_csv('data/medals_by_country_2016.csv', index_col=0)
````

````python
plt.style.use('default')
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of Medals")
plt.show()
# save 
fig.savefig("imgs/gold_medals.png", quality=50)
fig.savefig("imgs/gold_medals.png", dpi=100)

# change the size of the figure
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of Medals")
fig.set_size_inches([5,2])
plt.show()
````

### Autoamating Figures from Data
> One of the main strengths of Matplotlib is that it can be automated to adapt to the data that it receives as input.

````python
import pandas as pd

summer2016_medals = pd.read_csv('data/summer2016.csv')
summer2016_medals.sample()
````

````python
sports = summer2016_medals['Sport'].unique()

fig, ax = plt.subplots()
for sport in sports:
    sport_df = summer2016_medals[summer2016_medals['Sport']==sport]['Height']
    ax.bar(sport, sport_df.mean(), yerr=sport_df.std())
ax.set_xticklabels(sports, rotation=90)

fig.set_size_inches([8, 4])
plt.show()
````
