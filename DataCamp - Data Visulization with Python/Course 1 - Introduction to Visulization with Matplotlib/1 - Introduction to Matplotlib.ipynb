# Introduction to Matplotlib
- How to control visulizations with programs. 
- Many ways to use matplotlib, here we use object oriented interface called pyplot
```
fig, ax = plt.subplots()
```

- figure object: container that holds everything in the page

- ax object: canvas on which we draw the data

````python
import numpy as np
````

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(np.arange(10), np.random.rand(10))
ax.plot(np.arange(10), np.random.rand(10))
plt.show()
````

## Customizing the Plot
Title and labels with `ax.set_`

````python
fig, ax = plt.subplots()
x, y = np.arange(10), np.random.rand(10)
ax.plot(x, y, marker='o', linestyle="--", color="r")
# set x label
ax.set_xlabel("Time")
ax.set_ylabel("Magnitude")
ax.set_title("Ranomd Data")
plt.show()
````

## Small Multiples
multiple small subplots
```python
fig, ax = plt.subplots(rows, cols)
```
- this will create a grid of plots with rows and cols. We can access any subplot by index using `ax[row_num, col_number]`. 
- ax will be two dimensional if rows, cols>1 else it will be one dimensional. 


````python
fig, ax = plt.subplots(2,1)
x, y = np.arange(10), np.random.rand(10)
y1,y2 = np.random.rand(10)-1,np.random.rand(10)+1

# figure 0
ax[0].plot(x, y, marker='o', linestyle="--", color="r")
ax[0].plot(x, y1, marker='o', linestyle="--", color="b")
ax[0].plot(x, y2, marker='o', linestyle="--", color="b")
# one figure
ax[1].plot(x, y, marker='o', linestyle="--", color="r")
ax[1].plot(x, y1, marker='o', linestyle="--", color="b")
ax[1].plot(x, y2, marker='o', linestyle="--", color="b")
````

#### Subplots with Shared x or y axis
When subplots have a shared x-axis along a column, only the x tick labels of the bottom subplot are created. Similarly, when subplots have a shared y-axis along a row, only the y tick labels of the first column subplot are created. To later turn other subplots' ticklabels on, use tick_params.

````python
fig, ax = plt.subplots(2, 1, sharex=True)
x, y = np.arange(10), np.random.rand(10)
y1,y2 = np.random.rand(10)-1,np.random.rand(10)+1

# figure 0
ax[0].plot(x, y, marker='o', linestyle="--", color="r")
ax[0].plot(x, y1, marker='o', linestyle="--", color="b")
ax[0].plot(x, y2, marker='o', linestyle="--", color="b")
# one figure
ax[1].plot(x, y, marker='o', linestyle="--", color="r")
ax[1].plot(x, y1, marker='o', linestyle="--", color="b")
ax[1].plot(x, y2, marker='o', linestyle="--", color="b")
````
