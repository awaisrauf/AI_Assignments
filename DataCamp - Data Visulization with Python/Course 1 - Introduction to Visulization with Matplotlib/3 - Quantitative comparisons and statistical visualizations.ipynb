# Quantitative comparisons and statistical visualizations

```python
import pandas as pd
medals = pd.read_csv('data/medals_by_country_2016.csv', index_col=0)
```

## Bar Charts

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'], bottom=medals['Gold'], label='Gold')
ax.bar(medals.index, medals['Silver'], bottom=medals['Gold'], label='Silver')
ax.bar(medals.index, medals['Bronze'], bottom=medals['Gold']+medals['Silver'], label='Bronze')

ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Medals')
plt.legend()
plt.show()
```

## Histograms
- to show all the distribution of the data


```python
summer2016 = pd.read_csv('data/summer2016.csv')
mens_rowing = summer2016[summer2016['Sport']=='Rowing']
mens_gymnastics = summer2016[summer2016['Sport']=='Gymnastics']
```

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'], label='Rowing', bins=10, histtype='step')
ax.hist(mens_gymnastics['Height'], label='Gymnastics', bins=10, histtype='step')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')

ax.legend()
plt.show()
```

## Statistical Plotting
- adding error bars 

```python
fig, ax = plt.subplots()
ax.bar("Rowing", mens_rowing['Height'].mean(),
      yerr=mens_rowing['Height'].std(), label='Rowing')
ax.bar('Gymnastics', mens_gymnastics['Height'].mean(), 
      yerr=mens_gymnastics['Height'].std(), label='Gymnastics')

ax.set_ylabel('Height')
ax.legend()
plt.show()
```

### Error in Line Plot

```python
import numpy as np
x, y = np.arange(10), np.random.randn(10)
y_err = np.random.randn(10)

fig, ax = plt.subplots()
ax.errorbar(x, y, yerr=y_err)
```

### Boxplot

```python
fig, ax = plt.subplots()

ax.boxplot([mens_rowing['Height'], mens_gymnastics['Height']])
ax.set_xticklabels(['Rowing', 'Gymnastics'])
ax.set_ylabel('Height')
plt.show()
```

## Scatter Plots
- for Bivarite comparison 

```python
climate_data = pd.read_csv('data/climate_change.csv', index_col='date')
climate_data.index = pd.DatetimeIndex(climate_data.index)
```

```python
fig, ax = plt.subplots()
ax.scatter(climate_data['co2'], climate_data['relative_temp'])
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel('Relative temprature (Celsius)')
plt.show()
```

```python
eighties = climate_data['1980-01-01':'1989-12-31']
nineties = climate_data['1990-01-01':'1999-12-31']

fig, ax = plt.subplots()
ax.scatter(eighties['co2'], eighties['relative_temp'], label='eighties')
ax.scatter(nineties['co2'], nineties['relative_temp'], label='nineties')

ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel('Relative temprature (Celsius)')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(climate_data['co2'], climate_data['relative_temp'], c=climate_data.index)
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel('Relative temprature (Celsius)')
plt.show()
```
